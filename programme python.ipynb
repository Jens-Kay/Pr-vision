### Marchine Learning

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("C:/Users/maserati/Documents/Mémoire M1 - Copie/Base de données/base de données CSV/Csv/climsco_2021_1d_v30_depp_b3.csv", sep = ";")
df
```

```python
dfrandomtree = df[['CS_el', 'CS_peur', 'V_surnom', 'V_moq', 'V_ecart', 'V_malV', 'V_Bouscul', 'V_lancObj', 'V_frapp', 'V_menace', 'V_Bag', 'CS_toilettes_sec', 'VS_toilettes', 'VS_embrass', 'V_vols', 'V_Vol1', 'V_jeux2', 'pond_fin', 'indice_victimation', ]]
dfrandomtree
```

```python
dfrandomtree = dfrandomtree.fillna({'CS_el':5, 'CS_peur': 5,'V_surnom':5, 'V_moq' : 5,'V_ecart':5, 'V_malV':5, 'V_Bouscul':5, 'V_lancObj':5, 'V_frapp':5, 'V_menace':5, 'V_Bag':5, 'CS_toilettes_sec':5, 'VS_toilettes':5, 'VS_embrass':5, 'V_vols':3, 'V_Vol1':3, 'V_jeux2':4  })
```

```python
dfrandomtree.isna().sum()
```

```python
dfrandomtree['indice_vic_b'] = np.where(dfrandomtree['indice_victimation'] == 1, 0, 
                              np.where(dfrandomtree['indice_victimation'] == 2, 0,  
                                np.where(dfrandomtree['indice_victimation']== 3, 1,1)))
dfrandomtree
```

```python
dfrandomtree['V_Vols_b'] = np.where(dfrandomtree['V_vols']==2,1, np.where(dfrandomtree['V_vols']==1,0,3))
dfrandomtree['V_Vol1_b'] = np.where(dfrandomtree['V_Vol1']==2,1, np.where(dfrandomtree['V_Vol1']==1,0,3))
dfrandomtree
```

```python
X = dfrandomtree.drop(['V_vols', 'V_Vol1', 'pond_fin', 'indice_victimation', 'indice_vic_b'], axis=1) 
Y = dfrandomtree['indice_vic_b'] 
```

```python
for col in X.columns:  # Toutes les colonnes sauf la variable cible
    X[col] = X[col].astype('category').cat.codes + 1
X
```

```python
X['pond_fin'] = dfrandomtree['pond_fin']
```

## Création d'une table d'entraine et de test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
sample_weight = X_train['pond_fin'].values
sample_weight
```

# Arbre de décisions

```python
model = DecisionTreeClassifier(random_state=42)

model.fit(X_train, Y_train, sample_weight=sample_weight)


Y_pred = model.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)*100
print(f'Accuracy : {round(accuracy,2)}%')


cm = confusion_matrix(Y_test, Y_pred)

print("Matrice de Confusion :")
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f'Accuracy Arbre de décision : {round(accuracy,2)}%')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

# Forêts aléatoires (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X_train, Y_train, sample_weight=sample_weight)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, y_pred)*100

cm = confusion_matrix(Y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f'Accuracy Random Forest : {round(accuracy,2)}%')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

# Gradient Boosting Machines (GBM) 

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

gbm_model.fit(X_train, Y_train, sample_weight=sample_weight)

y_pred = gbm_model.predict(X_test)

accuracy = accuracy_score(Y_test, y_pred)*100


cm = confusion_matrix(Y_test, y_pred)


plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f'Accuracy GBM : {round(accuracy,2)}%')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = keras.Sequential([
    keras.layers.Dense(64, activation='sigmoid', input_shape=(X_train.shape[1],)),
    keras.layers.Dense(64, activation='sigmoid'),
    keras.layers.Dense(2, activation='softmax') 
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

H = model.fit(X_train, Y_train, epochs=50, sample_weight=sample_weight)

loss, accuracy = model.evaluate(X_test, Y_test)

y_pred_prob = model.predict(X_test)
y_pred = tf.argmax(y_pred_prob, axis=1)

accuracy = accuracy_score(Y_test, y_pred)*100


cm = confusion_matrix(y_test, y_pred)


plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f'Accuracy Neural Networks : {round(accuracy,2)}%')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
plt.plot(H.history['loss'], label='loss', marker='o')
plt.plot(H.history['accuracy'], label='accuracy', marker='o')

plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('loss par epoch')
plt.legend()

plt.show()
```

# K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


# Créer et entraîner un modèle KNN
knn_model = KNeighborsClassifier(n_neighbors=5)  # Choisissez le nombre de voisins
knn_model.fit(X_train, Y_train)

# Faire des prédictions
y_pred = knn_model.predict(X_test)


# Évaluer l'exactitude
accuracy = accuracy_score(Y_test, y_pred)*100

cm = confusion_matrix(Y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f'Accuracy KNN : {round(accuracy,2)}%')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

```
